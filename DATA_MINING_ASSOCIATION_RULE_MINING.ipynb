# 3. PRA-PEMROSESAN DATA



> A. MEMERIKSA INFORMASI

---





```python
# Menampilkan informasi dasar DataFrame
print("Informasi Dasar Dataset:")
data.info()
```



> B. MENGHITUNG MISSING VALUE

---






```python
print("Jumlah nilai kosong (NaN) per kolom:")
print(data.isnull().sum())
```



> C. MENANGANI TRANSAKSI KOSONG

---






MEMBERSIHKAN DATA DARI DATA KOSONG


```python
# Menyimpan jumlah transaksi awal
initial_count = len(data)

# Kita buat DataFrame baru bernama data_cleaned agar data asli tetap utuh.
data_cleaned = data.dropna(subset=['Items'])

# Mengecek string kosong atau hanya spasi
data_cleaned = data_cleaned[data_cleaned['Items'].str.strip() != '']

# Menyimpan jumlah transaksi setelah dibersihkan
cleaned_count = len(data_cleaned)
```


MENAMPILKAN HASIL




```python
# Menampilkan hasilnya
print(f"Jumlah transaksi awal: {initial_count}")
print(f"Jumlah transaksi setelah menghapus baris kosong: {cleaned_count}")
print(f"Jumlah transaksi yang dihapus: {initial_count - cleaned_count}")
```

MENGECEK ULANG APAKAH MASIH ADA DATA YANG KOSONG ATAU TIDAK

```python
# Memeriksa ulang ulang apakah masih ada nilai kosong di 'Items' pada data_cleaned
print("Jumlah nilai kosong di 'Items' setelah pembersihan:")
print(data_cleaned['Items'].isnull().sum())
```



> D. MENGUBAH STRING 'ITEMS' MENJADI LIST ITEM

---




MENERAPKAN FUNGSI SPLIT

```python
transactions_list = data_cleaned['Items'].apply(
    lambda x: [item.strip().lower() for item in x.split(', ')]
)
```

MENAMPILKAN DATA DALAM FORMAT LIST

```python
print("list item Transaksi:")
transactions_list.head(8)
```



> E. ENCODING TRANSAKSI KE FORMAT ONE-HOT ( MATRIX BINER )



PROSES FIT DAN TRANSFORM

```python
# Membuat objek TransactionEncoder
te = TransactionEncoder()

# Melakukan proses 'fit' dan 'transform' dalam satu langkah (fit_transform)
te_ary = te.fit_transform(transactions_list)

# Mengubah array boolean hasil encoding menjadi DataFrame Pandas
data_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

MENAMPILKAN HASIL

```python
# Menampilkan Hasil
print("Hasil DataFrame dalam format One-Hot Encoded :")
data_encoded.head(8)
```

```python
print("Daftar nama item unik (kolom):")
print(sorted(list(data_encoded.columns)))
```



> F. MENJALANKAN APRIORI UNTUK MENCARI FREQUENT ITEMSET




MENJALANKAN APRIORI DAN MENCATAT WAKTU EKSEKUSI


```python
# Tentukan nilai minimum support
min_support_apriori = 0.01
# Mencatat waktu mulai
start_time_apriori = time.time()

# Menjalankan algoritma Apriori
frequent_itemsets_apriori = apriori(data_encoded, min_support=min_support_apriori, use_colnames=True)
# Mencatat waktu selesai
end_time_apriori = time.time()
```

MENAMPILKAN HASIL

```python
# Menghitung dan tampilkan waktu eksekusi
execution_time_apriori = end_time_apriori - start_time_apriori
print(f"Algoritma Apriori selesai dalam {execution_time_apriori:.4f} detik.")

# Menampilkan jumlah frequent itemsets yang ditemukan
print(f"Jumlah frequent itemsets yang ditemukan: {len(frequent_itemsets_apriori)}")
```

```python
# Menampilkan beberapa frequent itemsets teratas
print("Contoh Frequent Itemsets (diurutkan berdasarkan support):")
frequent_itemsets_apriori.sort_values(by='support', ascending=False).head(10)
```

MEMBUAT ASSOCIATION RULES DARI FREQUENT ITEMSETS APRIORI MENGGUNAKAN LIFT

```python
# Menentukan metrik evaluasi dan threshold minimumnya
min_lift_threshold = 1.0
# Menghasilkan aturan asosiasi dari frequent itemsets yang ditemukan Apriori
rules_apriori = association_rules(frequent_itemsets_apriori, metric="lift", min_threshold=min_lift_threshold)
# Mengurutkan aturan berdasarkan 'lift' secara descending (dari tertinggi ke terendah)
rules_apriori_sorted = rules_apriori.sort_values(by='lift', ascending=False)
```

MENAMPILKAN HASIL

```python
# Menampilkan jumlah aturan yang ditemukan
print(f"Jumlah aturan asosiasi yang ditemukan dengan lift > {min_lift_threshold}: {len(rules_apriori_sorted)}")
# Menampilkan 10 aturan teratas berdasarkan lift
print("Top 10 Aturan Asosiasi berdasarkan Lift (dari Apriori):")
rules_apriori_sorted.head(10)
```



> G. MENJALANKAN FP-GROWTH UNTUK MENCARI FREQUENT ITEMSETS



MENJALANKAN ALGORITMA DAN MENCATAT WAKTU

```python
# Menggunakan min_support yang sama seperti Apriori untuk perbandingan yang adil
min_support_fpgrowth = 0.01
# Catat waktu mulai
start_time_fpgrowth = time.time()
# Jalankan algoritma FP-Growth
frequent_itemsets_fpgrowth = fpgrowth(data_encoded, min_support=min_support_fpgrowth, use_colnames=True)
# Catat waktu selesai
end_time_fpgrowth = time.time()
```

MENAMPILKAN HASIL

```python
# Menghitung dan Menampilkan waktu eksekusi
execution_time_fpgrowth = end_time_fpgrowth - start_time_fpgrowth
print(f"Algoritma FP-Growth selesai dalam {execution_time_fpgrowth:.4f} detik.")

# Bandingkan dengan waktu Apriori
print(f"Waktu eksekusi Apriori sebelumnya: {execution_time_apriori:.4f} detik.")
if execution_time_fpgrowth < execution_time_apriori:
    print("FP-Growth lebih cepat dari Apriori untuk dataset ini.")
elif execution_time_fpgrowth > execution_time_apriori:
    print("Apriori lebih cepat dari FP-Growth untuk dataset ini.")
else:
    print("Waktu eksekusi Apriori dan FP-Growth hampir sama.")
```

```python
print(f"Jumlah frequent itemsets yang ditemukan oleh FP-Growth: {len(frequent_itemsets_fpgrowth)}")
print(f"Jumlah frequent itemsets yang ditemukan oleh Apriori: {len(frequent_itemsets_apriori)}")

# Tampilkan beberapa frequent itemsets teratas dari FP-Growth (seharusnya mirip Apriori)
print("Frequent Itemsets dari FP-Growth (diurutkan berdasarkan support):")
frequent_itemsets_fpgrowth.sort_values(by='support', ascending=False).head(10)
```



> H. MENGHASILKAN ASSOCIATION RULES DARI FREQUENT ITEMSETS FP-GROWTH



MENGHASILKAN ATURAN ASOSIASI DAN KEMUDIAN MENGURUTKANNYA MENGGUNAKAN LIFT

```python
# Menghasilkan aturan asosiasi dari frequent itemsets yang ditemukan FP-Growth
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="lift", min_threshold=min_lift_threshold)

# Mengurutkan aturan berdasarkan 'lift' secara descending
rules_fpgrowth_sorted = rules_fpgrowth.sort_values(by='lift', ascending=False)
```

MENAMPILKAN HASIL

```python
# Menampilkan jumlah aturan yang ditemukan (seharusnya sama dengan Apriori)
print(f"Jumlah aturan asosiasi yang ditemukan dengan lift > {min_lift_threshold} (dari FP-Growth): {len(rules_fpgrowth_sorted)}")
print(f"Jumlah aturan asosiasi yang ditemukan oleh Apriori: {len(rules_apriori_sorted)}")
```

```python
# Menampilkan 10 aturan teratas berdasarkan lift dari FP-Growth
print("\nTop 10 Aturan Asosiasi berdasarkan Lift (dari FP-Growth):")
rules_fpgrowth_sorted.head(10)
```

# 4. INTERPRETASI HASIL



> A. INTERPRETASI SINGKAT



```python
# Tampilkan lagi 10 aturan teratas(Apriori dipilih karena lebih cepat dalam dataset ini)
print("Top 10 Aturan Asosiasi berdasarkan Lift (untuk diinterpretasi):")
# tampilkan kolom yang paling relevan untuk interpretasi
top_rules_display = rules_apriori_sorted[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
top_rules_display.head(10)
```

```python
print("--- Interpretasi Singkat Beberapa Aturan Teratas ---")

# interpretasi untuk aturan pertama
try:
    rule1 = top_rules_display.iloc[0]
    antecedents1 = ', '.join(list(rule1['antecedents']))
    consequents1 = ', '.join(list(rule1['consequents']))
    lift1 = rule1['lift']
    confidence1 = rule1['confidence']
    support1 = rule1['support']
    print(f"\n1. Aturan: Jika membeli {{{antecedents1}}}, maka cenderung membeli {{{consequents1}}}.")
    print(f"   - Lift = {lift1:.2f}: Pelanggan yang membeli '{antecedents1}' adalah {lift1:.2f} kali LEBIH MUNGKIN untuk juga membeli '{consequents1}' dibandingkan pelanggan pada umumnya.")
    print(f"   - Confidence = {confidence1:.2f}: Dari semua pelanggan yang membeli '{antecedents1}', sekitar {confidence1:.1%} juga membeli '{consequents1}'.")
    print(f"   - Support = {support1:.4f}: Kombinasi '{antecedents1}' dan '{consequents1}' muncul bersamaan dalam sekitar {support1:.2%} dari total transaksi.")

    # interpretasi untuk aturan kedua
    rule2 = top_rules_display.iloc[1]
    antecedents2 = ', '.join(list(rule2['antecedents']))
    consequents2 = ', '.join(list(rule2['consequents']))
    lift2 = rule2['lift']
    print(f"\n2. Aturan: Jika membeli {{{antecedents2}}}, maka cenderung membeli {{{consequents2}}}.")
    print(f"   - Lift = {lift2:.2f}: Pelanggan yang membeli '{antecedents2}' adalah {lift2:.2f} kali LEBIH MUNGKIN untuk juga membeli '{consequents2}'.")

except IndexError:
    print("\nTidak cukup aturan untuk menampilkan interpretasi contoh.")
except Exception as e:
    print(f"\nTerjadi error saat mencoba interpretasi: {e}")
```



> B. PERBANDINGAN KINERJA ALGORITMA ( WAKTU EKSEKUSI )



```python
print("--- Perbandingan Waktu Eksekusi Pencarian Frequent Itemsets ---")
print(f"Algoritma Apriori: {execution_time_apriori:.4f} detik")
print(f"Algoritma FP-Growth: {execution_time_fpgrowth:.4f} detik")

# Hitung perbedaan waktu
time_difference = abs(execution_time_apriori - execution_time_fpgrowth)
percentage_difference = (time_difference / execution_time_apriori) * 100 if execution_time_apriori > 0 else 0

if execution_time_fpgrowth < execution_time_apriori:
    print(f"\nHasil: FP-Growth lebih cepat sekitar {time_difference:.4f} detik ({percentage_difference:.2f}% lebih cepat).")
    print("Ini sesuai ekspektasi umum karena FP-Growth menghindari pembuatan kandidat berulang.")
elif execution_time_apriori < execution_time_fpgrowth:
     print(f"\nHasil: Apriori lebih cepat sekitar {time_difference:.4f} detik ({percentage_difference:.2f}% lebih cepat).")
     print("Ini kurang umum terjadi, mungkin karena datasetnya tidak terlalu besar atau overhead pembuatan FP-Tree lebih dominan.")
else:
    print("\nHasil: Waktu eksekusi kedua algoritma sangat mirip.")
```
